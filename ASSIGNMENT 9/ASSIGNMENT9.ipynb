```python
# ANS 1.


import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter






text = """Basketball is a fast-paced sport that requires skill, teamwork, and strategy. 
Players must dribble, pass, and shoot the ball to score points while defending against opponents. 
The game is played on a rectangular court with a hoop at each end. 
Professional leagues like the NBA have made basketball one of the most popular sports in the world. 
With exciting plays and intense competition, basketball continues to captivate fans globally."""


print("ANS 1.\n")

text = text.lower().translate(str.maketrans("", "", string.punctuation))
print("(i) "+text)
print("\n\n")
words = word_tokenize(text)
print(f"(ii) Words-> {words}\n")


sentences = sent_tokenize(text)
print(f"(ii) Sentences-> {sentences}\n\n")


stop_words = set(stopwords.words("english"))
filtered_words = [word for word in words if word not in stop_words]
print(f"(iii) {filtered_words}\n\n")


word_freq = Counter(filtered_words)
print(f"(iv) {word_freq}")



```

```python
# ANS 2.

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer
import string


print("ANS 2.\n")


print(f"(i) {filtered_words}\n\n")


porter = PorterStemmer()
lancaster = LancasterStemmer()
lemmatizer = WordNetLemmatizer()


porter_stems = [porter.stem(word) for word in filtered_words]
print(f"(ii) Poster Stem-> {porter_stems}\n")

lancaster_stems = [lancaster.stem(word) for word in filtered_words]
print(f"(ii) Lancaster Stem-> {lancaster_stems}\n\n")


lemmas = [lemmatizer.lemmatize(word) for word in filtered_words]
print(f"(iii) Lemmatizer-> {lemmas}\n\n")


print("(iv) Comparing the results: \n")
print("\n{:<15} {:<15} {:<15} {:<15}".format("Original", "PorterStemmer", "Lancaster", "Lemmatizer"))
print("-" * 60)
for o, p, l, le in zip(filtered_words, porter_stems, lancaster_stems, lemmas):
    print("{:<15} {:<15} {:<15} {:<15}".format(o, p, l, le))

```

```python
# ANS 3.

print("ANS 3.\n")


import re

print("2)\n")

words_5plus = re.findall(r'\b\w{6,}\b', text)
print(f"(a):{words_5plus}\n")



numbers = re.findall(r'\b\d+\b', text)
print(f"(b) {numbers if numbers else "No numbers found."}\n")


capitalized_words = re.findall(r'\b[A-Z][a-z]*\b', text)
print(f"(c) {capitalized_words}\n\n")


print("3)\n")
alpha_words = re.findall(r'\b[a-zA-Z]+\b', text)
print(f"(a) {alpha_words}\n")

vowel_words = re.findall(r'\b[aeiouAEIOU][a-zA-Z]*\b', text)
print(f"(b) {vowel_words} \n")

```

```python
# ANS 4.


import re


text = """Basketball is a fast-paced sport that requires skill, teamwork, and strategy.
Players must dribble, pass, and shoot the ball to score points while defending against opponents.
The game is played on a rectangular court with a hoop at each end.
Professional leagues like the NBA have made basketball one of the most popular sports in the world.
With exciting plays and intense competition, basketball continues to captivate fans globally."""
tl=text

def custom_tokenizer(text):
  
    tl = text.lower()

 
    pattern = r"""
        \b\w+(?:-\w+)+\b      # hyphenated words like fast-paced
        | \b\w+'\w+\b         # contractions like isn't, don't
        | \b\d+\.\d+\b        # decimal numbers like 3.14
        | \b\d+\b             # whole numbers like 100
        | \b[a-zA-Z]+\b       # regular words
    """

 
    tokens = re.findall(pattern, tl, re.VERBOSE)
    return tokens


tokens = custom_tokenizer(tl)
print("ANS 4. \n")


print(f"2. {tokens}\n\n")


text = re.sub(r'\b[\w\.-]+@[\w\.-]+\.\w+\b', '<EMAIL>', text)
text = re.sub(r'(https?://\S+|www\.\S+)', '<URL>', text)
text = re.sub(r'\b(?:\+91\s\d{10}|\d{3}-\d{3}-\d{4})\b', '<PHONE>', text)


print(f"3. {text}")
```
